## _**2a) Write a java program to print the nth value in the Fibonacci sequence in iterative and recursive ways.**_


```java
public class Fib {
  static long nthfib(int n) {
    if (n <= 2) {
      return 1;
    }
    int pre = 1, pst = 1;
    while (n != 2) {
      int s = pre + pst;
      pre = pst;
      pst = s;
      n--;
    }
    return pst;
  }
  static long nthfibr(int n) {
    if (n <= 2) {
      return 1;
    }
    return nthfibr(n - 1) + nthfibr(n - 2);
  }
  public static void main(String[] args) {
    System.out.println(nthfib(25));
    System.out.println(nthfibr(25));
  }
}

/*
Output:
75025
75025
*/
```

## _**2b) Write a java program to check whether the given matrix is symmetric or not.**_


```java
import java.util.Scanner;

public class Symmetric {

  public static void main(String[] args) {
    Scanner in = new Scanner(System.in);
    System.out.print("Enter size of square matrix: ");
    int n = in.nextInt();
    int[][] mat = new int[n][n];
    for (int i = 0; i < n; i++) {
      for (int j = 0; j < n; j++) {
        mat[i][j] = in.nextInt();
      }
    }
    boolean flag = true;
    for (int i = 0; i < n - 1; i++) {
      for (int j = 0; j < n - 1 - i; j++) {
        if (mat[i][j] != mat[j][i]) {
          flag = false;
        }
      }
    }
    if (flag) {
      System.out.println("Symmetric");
    }
    else {
      System.out.println("Non-symmetric");
    }
  }
}

/*
Output:
Enter size of square matrix: 3
1 2 3
2 4 5
3 5 6
Symmetric
*/
```

## _**3a) Write a java program that prompts the user for an integer n=2million and then prints out sum all prime numbers up to that integer.**_


```java
import java.util.Scanner;

public class SumPrime {

  public static void main(String[] args) {
    Scanner in = new Scanner(System.in);
    System.out.print("Enter n value: ");
    int n = in.nextInt();
    boolean[] prime = new boolean[n];
    for (int i = 2; i < n; i++) prime[i] = true;
    for (int p = 2; p * p < n; p++) {
      if (prime[p]) {
        for (int i = p * p; i < n; i += p)
          prime[i] = false;
      }
    }
    long sum = 0;
    for (int i = 3; i < n; i += 2) {
      if (prime[i]) sum += i;
    }
    System.out.println(
      "The sum of all prime numbers below " +n+": " + (sum + 2)
    );
  }
}

/*
Output:
Enter n value: 100
The sum of all prime numbers below 100: 1060
*/
```

## _**4a) Write a java program that reads a line of integers, and then display each integer, and the   sum of all the integers(use stringtokenizer class of java.util).**_


```java
import java.util.StringTokenizer;
import java.util.Scanner;

public class SumArr {

  public static void main(String[] args) {
    Scanner in = new Scanner(System.in);
    System.out.print("Enter Array: ");
    String s = in.nextLine();
    StringTokenizer st = new StringTokenizer(s, " ");
    int[] arr = new int[st.countTokens()];
    int i = 0, sum = 0;
    while (st.hasMoreTokens()) {
      arr[i++] = Integer.parseInt(st.nextToken());
    }
    System.out.print("Array: ");
    for (int j = 0; j < arr.length; j++) {
      System.out.print(arr[j] + " ");
      sum += arr[j];
    }
    System.out.println("\nSum of arr: " + sum);
  }
}

/*
Output:
Enter Array: 1 2 3 4
Array: 1 2 3 4
Sum of arr: 10
*/
```

## _**4b) Write a java program to implement Pascal triangle.**_


```java
import java.util.Scanner;

public class PascalTriangle {
  static int fact(int n) {
    if (n == 0) return 1;
    int res = 1;
    for (int i = 2; i <= n; i++) res = res * i;
    return res;
  }
  static int ncr(int n, int r) {
    return fact(n) / (fact(r) * fact(n - r));
  }
  public static void main(String[] args) {
    Scanner in = new Scanner(System.in);
    System.out.print("Enter n value: ");
    int n = in.nextInt();
    for (int x = 0; x <= n; x++) {
      System.out.print(" ".repeat(n-x));
      for (int y = 0; y <= x; y++) {
        System.out.print(ncr(x, y) + " ");
      }
      System.out.println();
    }
  }
}

/*
Output:
Enter n value: 4
    1
   1 1
  1 2 1
 1 3 3 1
1 4 6 4 1
*/
```

## _**5b) Write a java program to implement diamond shape.**_


```java
import java.util.Scanner;

public class Diamond {

  public static void main(String[] args) {
    Scanner in = new Scanner(System.in);
    System.out.print("Enter n value: ");
    int n = in.nextInt();
    for (int x = 1; x <= n; x++) {
      for (int y = 1; y <= n - x; y++) {
        System.out.print(" ");
      }
      for (int y = 1; y <= x * 2 - 1; y++) {
        System.out.print("*");
      }
      System.out.println();
    }
    for (int x = n-1; x>0; x--) {
      for (int y = 1; y <= n - x; y++) {
        System.out.print(" ");
      }
      for (int y = 1; y <= x * 2 - 1; y++) {
        System.out.print("*");
      }
      System.out.println();
    }
  }
}

/*
Output:
Enter n value: 4
   *
  ***
 *****
*******
 *****
  ***
   *
*/
```

## _**6b) Write a java program to find the sum of first n natural numbers using while loop.**_


```java
import java.util.Scanner;

public class Sumn {

  public static void main(String[] args) {
    Scanner in = new Scanner(System.in);
    System.out.print("Enter n value: ");
    int n = in.nextInt();
    int cn = n, sum = 0;
    while (cn > 0) {
      sum += cn;
      cn--;
    }
    System.out.println("Sum of " + n + " natural numbers: " + sum);
  }
}
/*
Enter n value: 10
Sum of 10 natural numbers: 55
*/
```

## _**7b)  Write a java program to find the sum of first n2 natural numbers using do-while loop.**_


```java
import java.util.Scanner;

public class Sumns {

  public static void main(String[] args) {
    Scanner in = new Scanner(System.in);
    System.out.print("Enter n value: ");
    int n = in.nextInt();
    int cn = n, sum = 0;
    do {
      sum += cn*cn;
      cn--;
    } while (cn > 0);
    System.out.println("Sum of " + n + "^2 natural numbers: " + sum);
  }
}
/*
Output:
Enter n value: 10
Sum of 10^2 natural numbers: 385
*/
```

## _**11b) Write a java program to find the sum of first n3 natural numbers using for each loop.**_


```java
import java.util.Scanner;
public class Sumn3 {

  public static void main(String[] args) {
    Scanner in = new Scanner(System.in);
    System.out.print("Enter n value: ");
    int n = in.nextInt();
    long sum = 0;
    int []arr = new int[n];
    for (int i = 0; i < n; i++) {
      arr[i] = i + 1;
    }
    for (int i : arr) {
      sum += i*i*i;
    }
    System.out.println(
      "Sum of the first n^3 numbers is: " + sum
    );
  }
}

/*
Output:
Enter n value: 5
Sum of the first n^3 numbers is: 225
*/
```

## _**12a) Write a java program to illustrate the concept of method overloading.**_


```java
public class MethodOverload {
  public static int add(int a, int b) {
    return a + b;
  }
  public static int add(int a, int b, int c) {
    return a + b + c;
  }
  public static void main(String[] args) {
    System.out.println("4 + 5 = " + add(4, 5));
    System.out.println("7 + 5 + 6 = " + add(7, 5, 6));
  }
}

/*
Output:
4 + 5 = 9
7 + 5 + 6 = 18
*/
```

## _**12b) Write a java program to check whether the given string is palindrome or not in optimal way.**_


```java
import java.util.Scanner;
public class Palindrome {

  public static void main(String[] args) {
    Scanner in = new Scanner(System.in);
    System.out.print("Enter String: ");
    String str = in.nextLine();
    String reverse = new StringBuffer(str).reverse().toString();
    if (str.equals(reverse)) {
      System.out.println(str + " is a palindrome.");
    } else {
      System.out.println(str + " is not a palindrome.");
    }
  }
}

/*
Output:
Enter String: bob
bob is a palindrome.
*/
```
